# 1. ARIMA time series model for BTC-USD 

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import joblib

# Load preprocessed data with additional features
btc_data = pd.read_csv('./Regression_Data/prep_BTC-USD_data.csv')

# Split data into training and testing sets (you can adjust the split ratio)
train_size = int(0.8 * len(btc_data))
train, test = btc_data.iloc[:train_size], btc_data.iloc[train_size:]

# Train a linear regression model on the features to predict 'Close'
regression_model = LinearRegression()
X_train = train[['Open', 'High', 'Low', 'Volume', 'Year', 'Month', 'Day']]
y_train = train['Close']
regression_model.fit(X_train, y_train)

# Predict 'Close' using the regression model
predicted_close = regression_model.predict(test[['Open', 'High', 'Low', 'Volume', 'Year', 'Month', 'Day']])

# Calculate residuals
residuals = test['Close'] - predicted_close

# Train ARIMA model on residuals
arima_model = ARIMA(residuals, order=(5, 1, 0))
fit_model = arima_model.fit()


# Save the trained model as a joblib file
model_filename = f"BTC-USD_arima_model.joblib"
joblib.dump(arima_model, model_filename)
print(f"Trained model saved as '{model_filename}'")
# Make predictions on the residuals
residual_predictions = fit_model.forecast(steps=len(test))

# Combine predicted residuals with regression predictions to get final predictions
final_predictions = predicted_close + residual_predictions

# Plot predicted vs. actual closing prices
plt.figure(figsize=(10, 6))
plt.plot(test.index, test['Close'], label='Actual Closing Price')
plt.plot(test.index, final_predictions, label='Predicted Closing Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('ARIMA Model with Exogenous Features Forecasting Closing Price')
plt.legend()
plt.show()
```
